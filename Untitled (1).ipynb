```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the function and its gradient
def f(x, y):
    return x + (x ** 2) + 2 * x * y + 2 * (y ** 2)

def grad_f(x, y):
    return 1 + 2 * x + 2 * y, 2 * x + 4 * y

# Perform gradient descent
def descente(x, y, l=1e-2, eps=1e-6, maxIter=1000):
    gradx, grady = grad_f(x, y)
    grad = math.sqrt(gradx ** 2 + grady ** 2)
    i = 0
    x_ = []
    y_ = []
    grad_ = []
    f_vals = []
    while abs(grad) > eps and i < maxIter:
        gradx, grady = grad_f(x, y)
        grad = math.sqrt(gradx ** 2 + grady ** 2)
        x = x - l * gradx
        y = y - l * grady
        i += 1
        grad_.append(grad)
        x_.append(x)
        y_.append(y)
        f_vals.append(f(x, y))
    return x, y, x_, y_, grad_, f_vals, i  # return number of iterations as well

# Test different learning rates
lambdas = np.linspace(0.01, 0.1, 10)  # Define a range of learning rates
iterations_list = []

for l in lambdas:
    result = descente(5, 6, l=l)
    iterations_list.append(result[6])

# Plot lambda vs. number of iterations
plt.figure()
plt.plot(lambdas, iterations_list, marker='o', linestyle='-', label='Iterations')
plt.xlabel('Learning Rate (λ)')
plt.ylabel('Number of Iterations')
plt.title('Learning Rate vs. Number of Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
